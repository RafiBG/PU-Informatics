# OpenAI, OllamaAPIs and Cloud providers for LLMs (DeepInfra, DeepSeek)

**[OpenAI](https://openai.com/)** е водеща компания в областта на изкуствения интелект, известна с разработката на модели като **GPT-4.5** и много други.
Техните модели се използват широко за задачи като генериране на текст, превод и разбиране на естествен език.
**OpenAI** предлага *API*, които позволяват на разработчиците да интегрират тези модели в своите приложения, предоставяйки мощни инструменти за обработка на език.​

**[DeepSeek](https://www.deepseek.com/)** е китайска компания, специализирана в разработката на големи езикови модели.
През ноември 2024 г. те представиха модела **DeepSeek-R1**, който е насочен към задачи, изискващи логически изводи, математически разсъждения и решаване на проблеми в реално време. Според някои източници, производителността на **DeepSeek-R1** е сравнима с тази на моделите на *OpenAI*, като същевременно е по-икономична. Това предизвика значителен интерес и дискусии в индустрията.

**[DeepInfra](https://deepinfra.com/)** предлага мащабируема и лесна за внедряване инфраструктура за машинно обучение, насочена към дълбоките езикови модели.
Тяхната платформа предоставя бърз и лесен достъп до водещи AI модели чрез опростен *API*, като се заплаща според използването.
Те предлагат различни модели, включително за генериране на текст, преобразуване на текст в реч и други, с акцент върху ниската латентност и ефективността на разходите.

**[Ollama API](https://ollama.com/)** е инструмент, който позволява локално изпълнение на големи езикови модели на нашата машина.
Той предоставя *REST API*, чрез който може да взаимодействат с тези модели. След като се инсталира Ollama и се изтегли желания модел, може да се използват *API* за генериране на отговори на дадени подканящи фрази.

Например, чрез командата:

`curl http://localhost:11434/api/generate -d '{ "model": "llama3.2", "prompt": "Земята плоска ли е?" }'`

`curl http://localhost:11434/api/generate \
  -H "Content-Type: application/json" \
  -d '{ "model": "llama2", "prompt": "Земята плоска ли е?" }'`

може да се получи отговор от модела. Тази функционалност позволява интеграция на LLM в различни приложения без необходимост от външни облачни услуги.
