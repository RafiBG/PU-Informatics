Вход – числото N, където 99999 < N < 1000000</h2>

Пример 1 </br>

Пример 2</br>

```python
num_as_str = input()

num_as_str_list = [x for x in num_as_str]

left_threesome = num_as_str[:3]
print(left_threesome)

right_thrreesome = num_as_str[-3:]
print(right_thrreesome)

left_threesome_sum = sum([int(x) for x in left_threesome])
print(left_threesome_sum)

right_threesome_sum = sum([int(x) for x in right_thrreesome])
print(right_threesome_sum)

if left_threesome_sum == right_threesome_sum:
  print(left_threesome_sum)
else:
  print(-1)
```

<h2>3. В машина може да се регулира температурата, като се задава исканата
стойност N и се избира един от следните два режима на работа:
R=1 – вдига температурата с 1 градус на минута.
R=2 – сваля температурата с 2 градуса на минута.
В момента, когато се достигне исканата температура, машината спира да
работи и я поддържа до задаване на нова.
След колко чáса и колко минути ще да се достигне зададената
температура?</h2>

Вход: Числата T-температурата в машината, N – желаната температура и R-
режима на работа.

Изход: две числа, разделени с интервал- час и минутата на достигане на
температурата.

Пример 1</br>
Вход</br>
20</br>
30</br>
1</br>
Изход</br>
0 10

Пример 2</br>
Вход</br>
20</br>
17</br>
2</br>
Изход</br>
0 2</br>

Пояснение на Пример 2: В края на първата минута температурата ще е 18
градуса, в края на втората вече ще е 17.

```python
current_temperature = int(input())
needed_temperature = int(input())
type = int(input())

hours = 0
minutes = 0

if current_temperature < needed_temperature and type == 1:

  difference = needed_temperature - current_temperature

  for num in range(difference):
    if minutes < 60:
      minutes += 1
    else:
      hours += 1

elif current_temperature > needed_temperature and type == 2:

  difference = needed_temperature - current_temperature

  for num in range(difference):
    if minutes < 120:
      minutes += 2
    else:
      hours += 1

print(f'{hours}, {minutes}')
```
