# Въведение в Markdown


Markdown е лек и популярен език за маркиране (Markup language), който се е наложил като стандард за писане на документи от изследователите и анализаторите на данни (data scientists and analysts). Често се преобразува в съответния HTML, чрез Markdown processor, което позволява лесното му споделяне между различни устройства и хора.

Езикът за маркиране (Markup language) е подобен на HTML (Hypertext Markup Language), направен от маркери за маркиране и се състои от отварящ <tagname> и затварящ маркер </tagname>.

* Заглавия (Headings)
* Блок цитати (Blockquotes)
* Секция с код (Code section)
* Математически символи (Mathematical Symbol)
* Нов ред (Line Break)
* Удебелен и курсивен текст (Bold and Italic Text)
* Хоризонтални линии (Horizontal Lines)
* Подреден списък (Ordered List)
* Неподреден списък (Unordered List)
* Таблица (Table)
* Изображение (Image)


## Заглавия

Заглавията започват с „# “, т.е. хеш символ, последван от интервала. Има шест заглавия с най-голямото заглавие, използващо само един хеш символ, а най-малките заглавия, използващи шест хеш символа:

\# (Header 1, title)

\## (Header 2, major headings)

\### (Header 3, subheadings)

\#### (Header 4)

\##### (Header 5)

\###### (Header 6)

Алтернативно, заглавията могат да започват с маркери за маркиране (Markup Tags) от h1 до h6 със следните синтаксиси:

\<h1\>Header 1,title\</h1\>

\<h2\>Header 2,major headings\</h2\>

\<h3\>Header 3,subheadings\</h3\>

\<h4\>Header 4\</h4\>

\<h5\>Header 5\</h5\>

\<h6\>Header 6\</h6\>


## Блок цитати

Блок цитатите съдържат големи части от текст и обикновено са отместени. Те могат да се използват чрез Markdown символа '>' или с \<blockquote\>Текст\</blockquote\>:

\>This is good

\<blockquote\>This is good\</blockquote\>

Резултат:



>This is good

## Секция с код

Секцията с код се използва за показване на код в различни програмни езици. Едната опция е да се сложи директно в текста, като се използва \`some inline code\` символ `n = 100`. Също може да се дефинира блок от код чрез използването на:

\```python

block of code

\```

Пример:

```python
a = 10
b = 30
print(a+b)
```

Алтернативно може да ползват маркери за маркиране \<code\>some code\</code\>“.

## Математически символи

Математически символи и изрази може да се използват в Markdown чрез изпозването на \$ символ, по следния начин \$mathematical symbols\$. Например да изведем корен квадратен от $1/k^2$ трябва да напишем \$\sqrt{\frac{1}{k^2}}\$:

$\sqrt{\frac{1}{k^2}}$

## Нов ред

Маркерът за прекъсване на редове е \<br\> без затварящ таг.

---

Следващия текст е написан на един ред:

Ред 1\<br\>Ред 2:

Ред 1<br>Ред 2


Използване на enter-a:

Line 1

Line 2

## Удебелен и курсивен текст

За удебеляване на текст може да ползвате \<b\> тагове или \*\* или \_\_ маркери:

\<b\>This is bold text \</b\>

\*\*This is bold text\*\*

\_\_This is bold text\_\_

За курсивен текст може да ползвате \<i\> тагове или \* или \_ маркери:

\<i\>This is italic text\</i\>

\*This is italic text\*

\_This is italic text\_

Примери:

**This is bold text**

*This is italic text*

## Хоризонтална линия

Хоризонтална линия в текста може да се добави чрез Markdown символи \-\-\- (три тирета) или Markup tag \<hr\>:

Текст над линията

---

Текст под линията

## Подреден списък

Подредения списък е номериран списък от редове:

1. One
2. Two
3. Three
  1. Four
  2. Five

В Markdown се дефинира чрез използването на "n. " в началото на всеки ред, където n е номера на реда:

\1. One

\2. Two

\3. Three

Алтернативно може да бъде дефиниран чрез отварящ и затварящ \<ol\> таг. Вътрешните редове се дефинират чрез \<li> тагове:

\<ol\>

\<li\>One\</li\>

\<li\>Two\</li\>

\<li\>Three\</li\>

\</ol\>

## Неподреден списък

Неподредените списък е списък от редове без номерация:

* One
* Two
* Three
- Four
  * Five

В Markdown се дефинира чрез използването на "- " или "* " в началото на всеки ред:

\* One

\* Two

\* Three

Алтернативно може да бъде дефиниран чрез отварящ и затварящ \<ul\> таг. Вътрешните редове се дефинират чрез \<li> тагове:

\<ul\>

\<li\>One\</li\>

\<li\>Two\</li\>

\<li\>Three\</li\>

\</ul\>

## Таблица



Таблици в Markdown се дефинират чрез използването на \| за разделител на колоните и \- за създаване на заглавна част и дефиниране на формата на клетките в таблицата:

---

\|Име|Адрес|Заплата|

\|-----|--------|------|

\|Иван|София|2000|

\|Петя|Пловдив|2500|

---

Горната дефиниция на таблица изглежда така:

|Име|Адрес|Заплата|
|--------|------------|-----------|
|Иван|София|2000|
|Петя|Пловдив|2500|
