1. Using list comprehension, construct a list from the squares of each element in the list.

```python
# Example:
list = [4, 5, 2, 3]

squares = [x ** 2 for x in list]
print(squares)
#result = [16,25,4,9]
```

2. Given dictionary is consisted of vehicles and their weights in kilograms. Using list comprehension, contruct a list of the names of vehicles in upper case with weight below 2000 kilograms.

```python
#Example:
vehicles = {"Sedan": 1500, "SUV": 2000, "Pickup": 2500, "Minivan": 1600, "Van": 2400, "Semi": 13600, "Bicycle": 7, "Motorcycle": 110}

light_vehicles = [vehicle.upper() for vehicle, weight in vehicles.items() if weight < 2000]
print(light_vehicles)
#result = ["SEDAN", "MINIVAN", "BICYCLE", "MOTORCYCLE"]
```

3. Slice the string until first "n".

```python
#example
str="Artificial intelligence"

position_n = str.find('n')
result = str[:position_n + 1]
print(result)
#print result should be "Artificial in"
```
