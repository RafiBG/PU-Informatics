# List comprehensions

```python
numbers = [1, 2, 3, 4, 5] # List comprehensions е синтактична конструкция за създаване на списъци по кратък и ефективен начин
squares = [x**2 for x in numbers]
print(f"Squares: {squares}")

# Filtering with list comprehensions
even_numbers = [x for x in numbers if x % 2 == 0]
print(f"Even numbers: {even_numbers}")

# Nested list comprehension
matrix = [[j for j in range(3)] for i in range(3)]
print(f"Matrix: {matrix}")
```

# Decorators

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()

# Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f"Hello, {name}!")

greet("Jason")
```

# Regex

```python
import re

# Pattern matching
pattern = r"\d+" # Match one or more digits
text = "There are 123 apples and 456 oranges."
matches = re.findall(pattern, text)
print(f"Matches: {matches}")

# Validate an email address
email = "example@domainname.com"
if re.match(r"[^@]+@[^@]+\.[^@]+", email): # RegEX
    print("Valid email address")
else:
    print("Invalid email address")

# Replace text
text = "I love Python programming."
replaced = re.sub(r"Python", "Java", text)
print(f"Replaced text: {replaced}")
```

# Dates times

```python
from datetime import datetime, timedelta

# Current date and time
now = datetime.now()
print(f"Current date and time: {now}")

# Formatting dates
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print(f"Formatted date: {formatted}")

# Adding and subtracting time
tomorrow = now + timedelta(days=1)
print(f"Tomorrow: {tomorrow}")

# Parsing a date string
date_string = "2025-01-01"
parsed_date = datetime.strptime(date_string, "%Y-%m-%d")
print(f"Parsed date: {parsed_date}")
```

# Advanced classes

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

class Student(Person):
    def __init__(self, name, age, major):
        super().__init__(name, age)
        self.major = major

    def greet(self):
        super().greet()
        print(f"I am studying {self.major}.")

student = Student("Allice", 22, "Computer Science")
student.greet()
```

# Multithreading

```python
import threading # Модулът threading позволява изпълнение на няколко нишки едновременно за постигане на многозадачност

def print_numbers():
    for i in range(1, 6):
        print(f"Number: {i}")

def print_letters():
    for letter in "ABCDE":
        print(f"Letter: {letter}")

thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```
