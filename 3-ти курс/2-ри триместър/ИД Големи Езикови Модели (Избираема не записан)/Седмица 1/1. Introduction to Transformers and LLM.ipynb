# Какво е голям езиков модел?

Големият езиков модел (LLM) е вид изкуствен интелект (AI), който може да генерира текст, да превежда изречения от един език на друг, да пише различни видове творческо съдържание и да отговаря на въпроси по информативен начин.

## Обучение на голям езиков модел.

LLM се обучават на масивни набори от данни, достъпни в интернет. „Large“ в големия езиков модел се отнася до броя на регулируемите параметри в модела. Колкото повече параметри има един модел, толкова по-сложен е той и от толкова повече данни може да се учи. Някои от най-успешните LLM имат стотици милиарди параметри, което им позволява да се учат от масивни набори от данни и да изпълняват сложни задачи.

Aналогия,  която може да помогне да се обясни тази концепция:

Нека си представим модел само с няколко параметъра. Този модел би бил като проста машина с няколко движещи се части. Може да изпълнява само прости задачи, като събиране на две числа.

А сега модел със стотици милиарди параметри. Този модел би бил като сложна машина с хиляди движещи се части. Може да изпълнява много по-сложни задачи, като превод на езици или писане на творческо съдържание.

Колкото повече параметри има един модел, толкова по-сложен е той и от толкова повече данни може да се учи. Това позволява на LLMs да изпълняват сложни задачи, които биха били невъзможни за по-малки модели.

# Какво е Transformer?

Повечето LLM са предварително обучени на голям набор от данни с общо предназначение и могат да бъдат допълнително прецизирани, за да се използват за конкретни случаи. Всичко това може да бъде възможно благодарение на трансформeрната архитектура.

image.png

Трансформeрите са тип архитектура на невронна мрежа, която се използва за задачи за обработка на естествен език (NLP). Те са представени за първи път в статията „[Attention is all you need](https://proceedings.neurips.cc/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf)“ от Vaswani et al. (2017).

Те работят, като използват механизми за самовнимание, за да научат дългосрочни зависимости в последователности. Това ги прави много по-мощни от традиционните архитектури на невронни мрежи, като повтарящи се невронни мрежи (RNN), за NLP задачи.

Архитектурата на трансформeра направи възможна широката адаптивност на LLM. Това направи обучението на моделите прогресивно в смисъл, че някой, който създава модел, може да не трябва да обучава модела напълно от нулата и може да използва предварително обучен модел, за да ги настрои фино, за да изпълняват специфичната си задача.

# Какво е фина настройка?

Фината настройка (fine-tune) е процес, при който предварително обучен голям езиков модел (LLM) се обучава допълнително върху по-малък, специфичен за задачата набор от данни. Това позволява на модела да научи специфичните характеристики на задачата и да подобри нейното изпълнение.

Примера за създаване на чатбот, който може да отговори на всички запитвания, свързани с един конкретен набор от документи. Нека се приеме, че тези документи са патентовани и не са достъпни в интернет.

image.png

Някой може да настрои финно съществуващ LLM чрез подаване на данни, специфични за тези документи и с някои повторения на обучение, моделът ще започне да отговаря на въпроси, свързани с тези документи.

Предимствата на фината настройка на LLM:

 * Може да подобри работата на LLM по конкретна задача.

 * Може да се използва за адаптиране на LLM към нов домейн или задача.

 * Може да се използва за подобряване на точността и надеждността на LLM.

Предизвикателствата при фината настройка на LLM:

 * Това може да отнеме много време и да е скъпо.

 * Може да е трудно да се намери специфичен за задача набор от данни, който е достатъчно голям и достатъчно висококачествен, за да обучи LLM.

 * Може да е трудно да се настроят хиперпараметрите на процеса на фина настройка.

## Архитектура на трансформeра:

Силата на трансформeрната архитектура се крие в нейната способност да научи уместността и контекста на всички думи в едно изречение.

image.png

## Токенизатори:

Архитектурата на трансформeра е разделена на две отделни части, енкодер и декодер. Тези компоненти работят във връзка един с друг и споделят редица прилики.
Моделите за машинно обучение са просто големи статистически калкулатори и работят с числа, а не с думи. Така че преди да се подадат текстове в модела за обработка, първо трябва да се токенизират думите с помощта на Tokenizers.

Това преобразува думите в числа, като всяко число представлява позиция в речник на всички възможни думи, с които моделът може да работи.

## Embeddings (вграждания):

Сега, когато входа е представен като числа, може да се предаде на слоя за вграждане.
Вграждането на векторни пространства се използва в обработката на естествен език. Алгоритмите за език от предишно поколение като Word2vec използват тази концепция.

В този прост случай всяка дума е съпоставена с идентификатор на токен и всеки токен е картографиран във вектор.

## СТЪПКИ, ЗА МИНАВАНЕ ПРЕЗ МОДЕЛА

Ще опростим работата на трансформерната архитектура за LLM и ще се види в 9 стъпки как тя прави работата на LLM такава, каквато е.

### Стъпка 1.
Преди да се подадат текстове в модела за обработка, първо трябва да се токенизират думите.

### Стъпка 2.
След като входът е представен като числа, може да се предадете на слоя за вграждане (ембединг). Този слой е пространство за вграждане на вектор, което може да се обучава, пространство с големи размери, където всеки токен е представен като вектор и заема уникално място в това пространство.

### Стъпка 3.
Докато се добавят символните вектори в основата на енкодера или декодера, също добавят позиционно кодиране. Чрез добавяне на позиционно кодиране се запазва информацията за словореда и не се губи уместността на позицията на думата в изречението.

### Стъпка 4.
След като се сумират входните токени и позиционните кодировки, се предават получените вектори на слоя за самовнимание (self-attention layer).

### Стъпка 5.
Теглата на самовниманието, които се научават по време на обучението и се съхраняват в тези слоеве, отразяват важността на всяка дума в тази входна последователност за всички други думи в последователността.

### Стъпка 6.
След като всички тегла на вниманието са приложени към нашите входни данни, изходът се обработва чрез напълно свързана мрежа за подаване напред (feed-forward).

### Стъпка 7.
Резултатът от този слой е вектор от логити (vector of logits), пропорционален на вероятностния резултат за всеки токен в речника на токенизатора.

### Стъпка 8.
След това тези логически стойности се подават на окончателния слой softmax*, където те се нормализират във вероятностен резултат за всяка дума. Този изход включва вероятност за всяка една дума в речника, така че тук вероятно ще има хиляди точки.

Softmax* е математическа функция, която преобразува вектор от необработени резултати (logits) от невронна мрежа във вероятностно разпределение, където всяка стойност е между 0 и 1 и общите суми до 1. Обикновено се използва в крайния слой на невронни мрежи за многокласови задачи за класификация, за да интерпретира изходните данни като вероятности за всеки клас.

### Стъпка 9.
Един единствен токен ще има по-висок резултат от останалите. Това е най-вероятният предвиден токен.
