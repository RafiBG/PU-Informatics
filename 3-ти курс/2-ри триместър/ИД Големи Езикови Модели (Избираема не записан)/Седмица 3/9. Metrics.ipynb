# Метрични показатели за класификация

При задачи за класификация, метричните показатели се използват за оценка на представянето на моделите.

Ето някои метрики:

* **Точност (Accuracy)**: Точността измерва процента на правилно класифицираните примери спрямо общия брой примери. Формулата за изчисляване на точността е:

Точност = $\frac{Брой \ правилно \ класифицирани \ примери}{Общ \ брой \ примери}$

Тя е полезна, когато класовете са балансирани, но може да бъде подвеждаща, ако класовете са неравномерно разпределени.

* **Прецизност (Precision)**: Прецизността измерва какъв процент от всички примери, които моделът е класифицирал като положителни, наистина са положителни.

Формулата е:

Прецизност = $\frac{Брой \ правилно \ предсказани \ положителни \ примери}{Брой \ всички \ реални \ положителни \ примери}$

Тя е важна, когато трябва да се избегнат фалшиви положителни класификации.

* **Извличане (Recall)**: Извличането измерва какъв процент от всички действителни положителни примери са били правилно класифицирани като положителни.

Формулата е:

Извличане = $\frac{Истински \ положителни}{Истински \ положителни \ + \ Фалшиви \ положителни}$

Тази метрика е полезна, когато е важно да не се изпуснат положителни примери, дори ако това води до повече фалшиви положителни.

* **F1-score**: F1 е средно хармонично на прецизността и извличането, и се използва, когато е важно да се балансират тези две метрики.

Формулата за F1 е:

F1-score = $2 * \frac{Прецизност \ * \ Извличане}{Прецизност \ + \ Извличане}$

Тя е полезна, когато има нужда от баланс между прецизността и извличането, особено при небалансирани класове.

# Weights & Biases ([W&B](https://wandb.ai/site))

Weights & Biases е платформа за управление на експерименти в машинното обучение. Тя предоставя инструментариум за проследяване, визуализиране и управление на моделите и резултатите от обучението.

Основните функции на W&B включват:

* **Проследяване на експерименти**: Позволява да се записват параметрите на модела, метриките по време на обучението и самите модели.

* **Визуализация на данни и метриките**: Може да се създават графики и визуализации, за да се анализира напредъка на обучението.

* **Управление на хиперпараметрите**: Лесно настройване и оптимизация на хиперпараметри.

* **Съхранение на модели и артефакти**: Може да се съхраняват и да се споделят модели и други артефакти от процеса на обучение.

    **Артефактите** обикновено включват всичко, което е резултат от процеса на обучение на модел или част от него: *химперпараметри, метрики, данни за обучение, графики и визуалиции, резултати от тестове* и други.

* **Колаборация и отчетност**: Позволява съвместна работа и проследяване на напредъка на различни екипи.

W&B е широко използван в индустриалната и академичната сфера за проследяване и оптимизиране на процесите на машинното обучение.

# [BGGlue](https://bgglue.github.io/)

BGGlue е набор от данни и еталон(бенчмарк) за естествено езиково разбиране на български език. Той е част от движението за развитие на инструменти и модели, специално адаптирани за българския език.
BGGlue съдържа различни задачи за оценка на модели по обработка на естествен език, включително:

* **Класификация на текстове**

* **Анализ на чувства**

* **Отговори на въпроси**

* **Извличане на информация**

Той е създаден с цел да подобри представянето на модели за обработка на естествен език, когато работят с български текстове, и да осигури стандартизиран начин за оценка на такива модели.
BGGlue е важен ресурс за изследователите и разработчиците, които работят с български език в контекста на машинното обучение.
