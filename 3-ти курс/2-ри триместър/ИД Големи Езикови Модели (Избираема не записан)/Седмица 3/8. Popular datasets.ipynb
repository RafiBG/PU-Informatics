# Известни набори от данни


Съществуват много популярни набори от данни в областта на обработката на естествен език (NLP) и машинното обучение, които се използват широко за различни задачи като класификация на текстове, отговори на въпроси, анализ на чувства и други.

Ето някои от тях:

1. **GLUE** (*General Language Understanding Evaluation*)

    Бенчмарк за оценка на представянето на модели при различни задачи на обработка на език.

    Включва задачи като сходство на изречения, извеждане на заключения и отговори на въпроси.

2. **SuperGLUE**

    Разширение на *GLUE* с по-трудни задачи, предназначено да затрудни моделите на ново ниво.

3. **CoNLL-2003**

    Набор от данни за задача за разпознаване на именовани обекти (*Named Entity Recognition* или *NER*), често използван за класификация на обекти в текстове.

4. **SQuAD** (*Stanford Question Answering Dataset*)

    Популярен набор от данни за отговори на въпроси, където моделите трябва да отговарят на въпроси въз основа на текстови пасажи създадени от *Crowdworkers*.

5. **RACE** (*ReAding Comprehension from Examinations*)

    Голям набор от данни за разбиране на текстове, съдържащ въпроси и затворени отговори за ученици от средно и гимназиално ниво.

6. **CLEF** (*Cross-Language Evaluation Forum*)

    Серия оценъчни кампании, целящи напредък в изследванията на многоезично извличане на информация и обработка на естествен език.

7. **EXAMS-BG**

    Набор от данни за български въпроси и отговори, свързани с българската образователна система.

8. **XLNI** (*Cross-lingual Natural Language Inference*)

    Набор от данни за многоезично извеждане на заключения от естествен език, където задачата е да се определи отношението между двойки изречения на различни езици.

9. **CINEXIO**

    Набор от данни за кино, включващ рецензии, сюжети и текстови данни, свързани с филми, често използван за задачи в NLP или задача за анализ на настроението.

10. **MNLI** (*Multi-Genre Natural Language Inference*)

    Голям набор от данни за извеждане на заключения от естествен език, където задачата е да се определи дали едно изречение предполага, противоречи или е неутрално спрямо друго.

11. **TREC** (*Text REtrieval Conference*)

    Набор от данни за класификация на въпроси, където задачата е да се категоризират въпросите в предварително дефинирани категории.

12. **WikiText**

    Голям набор от данни от английската Wikipedia, често използван за езиково моделиране.

13. **Amazon Reviews**

    Набор от данни за рецензии на продукти от Amazon, широко използван за анализ на чувства и класификация на текстове.

14. **Sentiment140**

    Набор от данни за анализ на чувства, съдържащ 1.6 милиона туита, маркирани с положително или отрицателно чувство.

15. **IMDB Reviews**

    Широко използван набор от данни за анализ на чувства, съдържащ рецензии на филми, маркирани като положителни или отрицателни.

16. **Toxic Comment Classification** (*Kaggle Dataset*)

    Набор от данни, предназначен за откриване на токсични/негативни коментари в онлайн дискусии.

17. **Common Crawl**

    Огромен набор от данни, съдържащ уеб данни, който може да се използва за различни задачи в NLP като езиково моделиране, разпознаване на обекти и др.

18. **Flickr30k**

    Набор от данни, съдържащ 30 000 изображения с асоциирани надписи, често използван за задачи като генериране на надписи за изображения.

19. **MS COCO** (*Microsoft Common Objects in Context*)

    Набор от данни с изображения и надписи за задачи като генериране на надписи за изображения, разпознаване на обекти и сегментиране.

20. **ParaNMT-50M**

    Набор от данни за машинен превод, особено фокусиран върху двойки изречения за 50 езика.

21. **Yelp Reviews**

    Голям набор от данни с рецензии от *Yelp*, често използван за анализ на чувства и класификация на текстове.

22. **OpenSubtitles**

    Огромна колекция от субтитри за филми и телевизионни шоута, често използвана за задачи като генериране на текст, езиково моделиране и машинен превод.

Тези набори от данни са широко използвани в научната област за обучение и оценка на NLP модели като играят важна роля в изследванията и развитието на нови алгоритми и модели.
