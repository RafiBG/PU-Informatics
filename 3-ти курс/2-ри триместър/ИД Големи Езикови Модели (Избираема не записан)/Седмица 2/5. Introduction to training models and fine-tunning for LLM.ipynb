# Pretraining VS. Fine-tuning

Езиковите модели са направили значителни крачки в разбирането и генерирането на естествен език. Тук ще говорим за непрекъснатото предварително обучение и ще се подчертаят разликите му от стратегиите за фина настройка. Ще се покажат различни примери за този иновативен подход, който притежава потенциала да промени бъдещето на езиковите модели.

Нека видим разликата между фина настройка и непрекъснато предварително обучение. Те може да звучат подобно, но са два различни процеса, често погрешно наричани просто „фина настройка“.

LLM са показали задълбочено разбиране на естествения език, подобрявайки представянето на набор от задачи. Използването на отворени уеб данни помогна за създаването на LLM с общо предназначение с широк набор от възможности. LLM с общо предназначение обаче не са „специалисти“; например, докато LLM могат да пишат добри новинарски статии, би било трудно да пишат специализирани правни документи.

Най-често се говори за фина настройка, когато се вземе предварително обучен модел и се адаптира за конкретна задача (като класификация или отговаряне на въпроси) или набор от данни. Ако обаче се установи, че този целеви набор от данни е от конкретен домейн и има няколко немаркирани данни, които биха могли да помогнат на модела да се адаптира към този конкретен домейн, тогава може да се направи нещо, наречено MLM или MLM+NSP „фина настройка“ (известна още като непрекъснато или допълнително предварително обучение).

За да се създадат специфични за домейн езикови модели, има два основни подхода:

 * Обучение от нулата: Изграждане на специализирани LLM изцяло от данни, специфични за домейна.
 * Непрекъснато предварително обучение: Подобряване на съществуващите LLM чрез допълнителното им предварително обучение с данни, специфични за домейна.

Тези два метода позволяват ефективно LLM да се приспособи към конкретни домейни.

Наистина има разлика между фина настройка за конкретна задача и придобиване на знания, специфични за домейна.

# Какво е Pretraining?

Pre-training обикновено означава вземане на оригиналния модел, инициализиране на теглата на случаен принцип и обучение на модела от нулата върху някои големи корпуси.

Допълнително/непрекъснато (Further/Continuous) pre-training означава да се вземе някакъв вече обучен модел и основно да се приложи трансферно обучение — използвайки вече запазените тегла от обучения модел (контролна точка) и да се обучи върху някакъв нов домейн (напр. финансови данни).

image.png

Както е показано на снимката, непрекъснатото предварително обучение разчита на концепцията за трансферно обучение. След като моделът е преминал първоначално предварително обучение, той може да приложи своите научени езикови модели към нови набори от данни. Този подход използва немаркирани данни от конкретна област, позволявайки на езиковия модел да подобри своето разбиране и ефективност в специфични области на знания, като финанси, право или здравеопазване.

Както можем да се види в таблицата, версиите на модела за продължаващо предварително обучение FinPythia-6.9B и FinPythia-1B показват превъзходна производителност при задачи FPB, Headline и NER, като същевременно показват сравнително по-ниски резултати при задачата FiQA SA в сравнение с Pythia. [For more info](https://arxiv.org/abs/2311.08545).

Предварителното обучение е първият етап от обучението на **голям езиков модел** (**LLM**) върху обширен корпус от текстови данни (.csv или .json). Моделът се учи чрез **обучение без надзор**, предвиждайки следващата дума в изречението. Тази фаза позволява на модела да разбере **граматиката, синтаксиса, семантиката и общите познания за света**. Предварителното обучение изисква **огромни набори от данни**, често извадени от интернет, книги и статии. Това е изчислително интензивно и обикновено се изпълнява на **високопроизводителни GPU** или **TPU**. След предварително обучение моделът има общи познания, но му липсва специализация за конкретни задачи.

Oбучение без надзор (неконтролирано обучение): Неконтролираното обучение е вид **машинно обучение**, при което модел научава модели от **немаркирани данни**. Той идентифицира **скрити структури, клъстери или връзки** без изрични инструкции. Общите техники включват **групиране** (**напр. K-Means, DBSCAN** (клъстери)) и **намаляване на размерността** (**напр. PCA** (компресия), **t-SNE** (визуализация на високоизмерни данни в 2D/3D)). Този подход се използва при откриване на **аномалии, системи за препоръки и LLMs за предварително обучение**. Моделът се учи чрез намиране на модели и **прилики в данните**.

# Какво е Fine-tuning?

Фината настройка използва етикетирани данни за фина настройка на параметрите на модела, като го приспособява към специфичните нюанси на дадена задача. Тази специализация значително повишава ефективността на модела в тази конкретна задача в сравнение с предварително обучен модел с общо предназначение.

image.png

Alpaca и Vicuna са фино настроени версии на модела LLaMA с възможност за участие в разговори и следване на инструкции. Следователно се очаква поведението му да прилича на това на ChatGPT.

Но колко добри са те? Според техния уебсайт качеството на изхода на Vicuna (според GPT-4…) е около 90% от ChatGPT, което го прави най-добрият езиков модел, който можем да изпълняваме локално. Това означава, че чрез фина настройка на модел можем да получим много по-добра версия на базирания модел за конкретна задача.

image.png

----------------------------------------

Фината настройка е процес, при който п**редварително обучен модел** се обучава допълнително върху **конкретен набор от данни**, за да се специализира в определени задачи. Той често използва **контролирано обучение**, където етикетираните данни насочват модела към по-добра точност. Фината настройка може да помогне за адаптирането на модела към области като **медицина, право или финанси**, което го прави по-прецизен в тези области. В някои случаи за привеждане в съответствие се използват техники като **обучение с подсилване с човешка обратна връзка** (**RLHF**). Този процес изисква по-малко данни и изчислителна мощност в сравнение с предварителното обучение. Добре настроеният модел запазва общи познания, като същевременно се отличава в целевата си област.

Oбучение с надзор (контролирано обучение): Контролираното обучение е метод за **машинно обучение**, при който моделът се учи от **етикетирани данни** (labeled data). Всеки вход има съответен **правилен изход**, което позволява на модела да научи изрично модели. Общите алгоритми включват **дървета на решенията, невронни мрежи и поддържащи векторни машини** (**SVM**). Този подход се използва при **класификация** (**напр. откриване на спам**) **и регресия** (**напр. прогнозиране на цените**). Моделът се подобрява чрез минимизиране на **грешките между прогнозите и действителните резултати**, използвайки обратна връзка от етикетирани примери.

image.png

В обобщение, езиковите модели (LLM) придобиват знания чрез комбинация от предварително обучение и фина настройка. Фината настройка е от съществено значение за приспособяването на модели към конкретни задачи като обобщаване, въпроси и отговори или класификация. Междувременно непрекъснатото предварително обучение подобрява LLMs с по-задълбочени познания в области като медицина, финанси или право. В зависимост от нашия случай на употреба можете да използвате фина настройка, непрекъснато предварително обучение или и двата подхода за оптимални резултати.

# Model training 🧪

Зареждане на модела.

1. Max Sequence Length: Максимална дължина на последователност `max_seq_length 5020`, максималният брой токени, които могат да бъдат използвани в модела, който може да обработва в една входна последователност. Това е от решаващо значение за задачи, изискващи обработка на дълги текстове, като се гарантира, че моделът може да улови повече контекст при всяко преминаване.
2.  Зареждане на Llama 3.2: Моделът и токенизаторът се зареждат с помощта на `FastLanguageModel.from_pretrained` със специфичен предварително обучен модел, `"unsloth/Llama-3.2-1B-bnb-4bitt"`. Това е оптимизирано за 4-битова точност, което намалява използването на паметта и увеличава скоростта на обучение, без значително да компрометира производителността. Параметърът `load_in_4bit=True` позволява това ефективно 4-битово квантуване (quantization), което го прави по-подходящ за фина настройка на по-малко мощен хардуер.
3. Прилагане на PEFT (Parameter-Efficient Fine-Tuning): След това се конфигурира модела с помощта на `get_peft_model`, който прилага **LoRA** (Low-Rank Adaptation) техники. Този подход се фокусира върху фина настройка само на конкретни слоеве или части от модела, а не на цялата мрежа, драстично намалявайки необходимите изчислителни ресурси.

 Параметри като `r=16` и `lora_alpha=16` регулират сложността и мащаба на тези адаптации. Използването на `target_modules` указва кои слоеве на модела трябва да бъдат адаптирани, които включват ключови компоненти, участващи в механизми за внимание като `q_proj`, `k_proj` и `v_proj`.

 `use_rslora=True` активира стабилизатора по ранг LoRA, което подобрява стабилността на процеса на фината настройка. `use_gradient_checkpointing="unsloth"` гарантира, че използването на паметта е оптимизирано по време на обучение чрез селективно съхраняване само на необходимите изчисления, което допълнително повишава ефективността на модела.
4. Verifying Trainable Parameters: Проверка на подлежащите на обучение параметри използва `model.print_trainable_parameters()`, за да отпечата броя на параметрите, които ще бъдат актуализирани по време на фината настройка, което позволява да се провери дали само предвидените части от модела се обучават.

Тази комбинация от техники прави процеса на фината настройка не само по-ефективен, но и по-достъпен, което позволява да се внедри този модел дори с ограничени изчислителни ресурси.

 Задаването на максимална дължина на `tokenz 5020` е повече от достатъчно като LoRA за обучение.


```python
max_seq_length = 512 # 5020;

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Llama-3.2-1B-bnb-4bit",
    max_seq_length=max_seq_length,
    load_in_4bit=True,
    dtype=None, # Позволява автоматичен избор на типа данни (FP16, BF16 и т.н.), в зависимост от хардуера
)

model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=["q_proj", "k_proj", "v_proj", "up_proj", "down_proj", "o_proj", "gate_proj"],
    use_rslora=True,
    use_gradient_checkpointing="unsloth",
    random_state = 32,
    loftq_config = None,
)

print(model.print_trainable_parameters())
```

* Data Prompt Structure: Структура на промпт за данни `data_prompt` е шаблон за форматиран низ, предназначен да насочва модела при анализирането на предоставения текст. Той включва контейнери за въведения текст (контекста) и отговора на модела. Този шаблон специално подканва модела да идентифицира индикатори за психично здраве, което улеснява фината настройка на модела за задачи, свързани с психичното здраве.

* End-of-Sequence Token: Токен за край на последователността `EOS_TOKEN` се извлича от токенизатора, за да обозначи края на всяка текстова последователност. Този токен е от съществено значение за модела, за да разпознае кога подкана е приключила, като помага да се поддържа структурата на данните по време на обучение или извод.

* Formatting Function: Функция за форматиране `Formatting_prompt` се използва за вземане на пакет от примери и ги форматира според `data_prompt`. Той итерира входните и изходните двойки, като ги вмъква в шаблона и добавя `EOS` токена в края. След това функцията връща речник, съдържащ форматирания текст, готов за обучение или оценка на модела.

* Изход на функцията: Функцията извежда речник, където ключът е **"текст"**, а стойността е списък от форматирани низове. Всеки низ представлява напълно подготвена подкана за модела, съчетаваща контекста, отговора и структурирания шаблон на подкана.

```python
data_prompt = """Analyze the provided text from a mental health perspective. Identify any indicators of emotional distress, coping mechanisms, or psychological well-being. Highlight any potential concerns or positive aspects related to mental health, and provide a brief explanation for each observation.

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompt(examples):
    inputs = examples["Context"]
    outputs = examples["Response"]
    texts = []
    for input_, output in zip(inputs, outputs):
        text = data_prompt.format(input_, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
```

Форматиране данните за обучение.

```python
training_data = Dataset.from_pandas(filtered_data)
#training_data = training_data.shuffle(seed=42).select(range(150)) # Train on only 150 samples not all

training_data = training_data.map(formatting_prompt, batched=True)
```

# Настройка на trainer за започване на фина настройка!

 1. Инициализация на trainer: Ще се инициализира `SFTTrainer` с модела и токенизатора, както и набора от данни за обучение. Параметърът `dataset_text_field` указва полето в набора от данни, което съдържа текста, който да се използва за обучение. Обучителят е отговорен за управлението на процеса на фината настройка, включително обработка на данни и актуализации на модела.
 2. Training Arguments: Аргументи за обучение, класът `TrainingArguments` се използва за дефиниране на ключови хиперпараметри за процеса на обучение. Те включват:

  - `learning_rate=3e-4`: Задава скоростта на обучение за оптимизатора.
  - `per_device_train_batch_size=32`: Определя batch size на устройството, оптимизирайки използването на **GPU**.
  - `num_train_epochs=20`: Указва броя на тренировъчните епохи.
  - `fp16=not is_bfloat16_supported()` и `bf16=is_bfloat16_supported()`: Активира смесено прецизно обучение, за да намали използването на паметта, в зависимост от хардуерната поддръжка.
  - `optim="adamw_8bit"`: Използва 8-битовия оптимизатор на **AdamW** за ефективно използване на паметта.
  - `weight_decay=0.01`: Прилага намаляване на теглото, за да се предотврати пренастройването.
  - `output_dir="output"`: Указва директорията, където ще бъдат записани обученият модел и регистрационните файлове.
  3.  Процес на обучение: Накрая се извика метода `trainer.train()`, за да стартира процеса на обучение. Той използва дефинираните параметри на фината настройка на модела, коригиране на тегла и обучение от предоставения набор от данни.
```python
text="""I'm going through some things with my feelings and myself. I barely sleep and I do nothing but think about how I'm worthless and how I shouldn't be here. I've never tried or contemplated suicide. I've always wanted to fix my issues, but I never get around to it. How can I change my feeling of being worthless to everyone?"""
```

`model = FastLanguageModel.for_inference(model)` конфигурира модела специално за извод, като оптимизира неговата производителност за генериране на отговори.

 Входящият текст се токенизира с помощта на `tokenizer`, той преобразува текста във формат, който моделът може да обработва. Използва се `data_prompt` за форматиране на входния текст, докато контейнерът за отговор остава празен, за да се получи отговор от модела. Параметърът `return_tensors = "pt"` указва, че изходът трябва да бъде в **тензори** на **PyTorch**, които след това се преместват в **GPU** с помощта на `.to("cuda")` за по-бърза обработка.

 `model.generate` генерира отговор въз основа на токенизираните входове. Параметрите `max_new_tokens = 5020` и `use_cache = True` гарантират, че моделът може да произвежда дълги и съгласувани отговори ефективно чрез използване на кеширани изчисления от предишни слоеве.

```python
model = FastLanguageModel.for_inference(model)

inputs = tokenizer(
[
    data_prompt.format(
        #instructions
        text,
        #answer
        "",
    )
], return_tensors = "pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens = 5020, use_cache = True)
answer=tokenizer.batch_decode(outputs)
answer = answer[0].split("### Response:")[-1]

print("Answer:", answer)
```
